```python
import pandas as pd        # import the Python Data Analysis Library with the shortened name pd
df = pd.read_csv("BP.CSV") # read in the file into a pandas dataframe
df
```

```python
df.info()
```

```python
df=df.drop("compound", axis=1)
df
```

```python
# Remove columns with object dtype
df = df.select_dtypes(exclude=['object'])
df
```

```python
import pandas as pd
import numpy as np

# Calculate z-scores for each column
z_scores = np.abs((df - df.mean()) / df.std())

# Define a threshold for outlier detection
threshold = 3

# Remove rows with any z-score greater than the threshold
df = df[(z_scores < threshold).all(axis=1)]

# Print the resulting DataFrame without outliers
df
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Select the correlation values for "BP_K" column
correlation_with_BP_K = correlation_matrix["BP_K"]

# Print the correlation values for "BP_K" column
print(correlation_with_BP_K)
```
